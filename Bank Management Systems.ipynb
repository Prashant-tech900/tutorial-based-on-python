```python
import pymysql as pq

def create_data_base():
    global conn, cu
    conn = pq.connect(
        host='localhost',
        user='root',
        password=''
    )
    query = ''' CREATE DATABASE IF NOT EXISTS wscube_bank_database'''
    cu = conn.cursor()
    cu.execute(query)


def connect_to_data_base():
    global conn, cu
    conn = pq.connect(
        host='localhost',
        user='root',
        password='',
        database='wscube_bank_database'
    )
    cu = conn.cursor()


def create_table_in_database():

    query = '''CREATE TABLE IF NOT EXISTS wscube_bank_table(
    name varchar(255) NOT NULL,
    account_no INT PRIMARY KEY AUTO_INCREMENT, 
    password varchar(255) NOT NULL,
    account_balance float ,
    email varchar(255),
    phone_no BIGINT NOT NULL,
    CONSTRAINT check_name_not_empty CHECK(name <> ''),
    CONSTRAINT check_password_not_empty CHECK(password <> ''),
    CONSTRAINT check_phone_no_not_empty CHECK(phone_no <> 0)
    );'''

    cu = conn.cursor()
    cu.execute(query)

    query = ''' ALTER TABLE wscube_bank_table AUTO_INCREMENT = 1000000; '''
    cu = conn.cursor()
    cu.execute(query)


def insert_into_wscube_bank_table(name, password, account_balance, email, phone_no):
    try:
        query = "INSERT INTO wscube_bank_table (name, account_no, password, account_balance, email, phone_no) VALUES (%s, NULL, %s, %s, %s, %s)"
        values = (name, password, account_balance, email, phone_no)
        cu.execute(query, values)
        conn.commit()
        print("Your account has been succesuflly created!")

    except conn.Error as error:
        print("Error while inserting data into the table:", error)


# This function is needed only when user is creating new account because account no.
# get generated after  inserting data and generating account no.
def fetch_account_no_from_database(phone_no):

    query = f"SELECT account_no FROM wscube_bank_table WHERE phone_no = '{phone_no}'"

    cu.execute(query)
    row = cu.fetchone()
    account_no = row[0]
    return account_no


def fetch_data_from_database(account_no, what_to_fetch=None):
    ''' This function takes have only 1 key word argument "what_to_fetch"
         that can get following attributes [name, password, account_no, phone_no, account_balance] '''

    if what_to_fetch == None:
        return None

    elif what_to_fetch == "name":
        query = f"SELECT name FROM wscube_bank_table WHERE account_no = '{account_no}'  "

    elif what_to_fetch == "password":
        query = f"SELECT password FROM wscube_bank_table WHERE account_no = '{account_no}'  "

    elif what_to_fetch == "phone_no":
        query = f"SELECT phone_no FROM wscube_bank_table WHERE account_no = '{account_no}'  "

    elif what_to_fetch == "account_balance":
        query = f"SELECT account_balance FROM wscube_bank_table WHERE account_no = '{account_no}'  "

    elif what_to_fetch == "email":
        query = f"SELECT email FROM wscube_bank_table WHERE account_no = '{account_no}'  "

    elif what_to_fetch == "account_no":
        query = f"SELECT account_no FROM wscube_bank_table WHERE account_no = '{account_no}'  "

    cu.execute(query)
    row = cu.fetchone()
    data = row[0]
    return data


def update_data_in_database(account_no, what_to_update=None, new_data=None):

    if what_to_update == None:
        return None

    elif what_to_update == "name":
        query = f''' UPDATE `wscube_bank_table` SET `name` = '{new_data}' where account_no = '{account_no}'  '''

    elif what_to_update == "password":
        query = f''' UPDATE `wscube_bank_table` SET `password` = '{new_data}' where account_no = '{account_no}'  '''

    elif what_to_update == "phone_no":
        query = f''' UPDATE `wscube_bank_table` SET `phone_no` = '{new_data}' where account_no = '{account_no}'  '''

    elif what_to_update == "email":
        query = f''' UPDATE `wscube_bank_table` SET `email` = '{new_data}' where account_no = '{account_no}'  '''

    elif what_to_update == "account_balance":
        query = f''' UPDATE `wscube_bank_table` SET `account_balance` = '{new_data}' where account_no = '{account_no}'  '''

    cu.execute(query)
    conn.commit()


def account_exists(account_no):

    query = ''' SELECT account_no from wscube_bank_table '''
    cu.execute(query)
    x = cu.fetchall()
    for rows in x:
        for account_nos in rows:
            if int(account_no) == account_nos:
                return True
    else:
        return False
```
